```python
import pandas as pd

# Sample DataFrame
data = {
    'Gender': ['Male', 'Female', 'Male', None, 'Female']
}

df = pd.DataFrame(data)

# Convert 'Gender' to boolean: 'Male' becomes True, 'Female' becomes False, None becomes False
df['Gender'] = df['Gender'].map({'Male': True, 'Female': False}).astype('boolean')  # Use 'boolean' to handle None values

# Print the modified DataFrame and its datatypes
print(df)
print(df.dtypes)
```

```python
# Check if the column exists before attempting to delete it
if 'Family_Income' in df.columns:
    del df['Family_Income']  # Delete the column
    print("\nColumn 'Family_Income' deleted.")
else:
    print("\nColumn 'Family_Income' does not exist in the DataFrame.")

# Display the DataFrame after deletion
print("\nDataFrame after deletion:")
print(df)
```

```python
import matplotlib.pyplot as plt

# Assuming you have 'Student_Name' and 'Score' columns in your DataFrame
# Replace 'Score' with the actual column name that contains the exam scores
df = pd.read_csv("C:\\Users\\SEBALAMAKGOLO3\\Desktop\\StudentPerformanceFactors.csv") #File path

plt.bar(df['Hours_Studied'], df['Exam_Score'], color='pink')  # Use 'Student_Name' for x-axis and 'Score' for y-axis
plt.xlabel('Hours_Studied')
plt.ylabel('Exam_Score')
plt.title('Students Exam Progress')
plt.xticks(rotation=45)  # Rotate names for better visibility
plt.tight_layout()  # Adjust layout to prevent clipping
plt.grid()
plt.show()  # Display the chart
```

```python
import matplotlib.pyplot as plt
 
# Define the data
labels = ['Min (55)', '25% (65)', '50% (67)', '75% (69)', 'Max (101)']
sizes = [55, 65, 67, 69, 101]  # The scores
colors = ['gold', 'lightcoral', 'lightskyblue', 'lightgreen', 'lightpink']
explode = (0.1, 0, 0, 0, 0)  # explode the 1st slice (Min)
 
# Create the pie chart
plt.figure(figsize=(8, 8))
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=140)
 
# Equal aspect ratio ensures that pie chart is a circle.
plt.axis('equal')  
plt.title('Exam Scores Distribution')
plt.show()
```
